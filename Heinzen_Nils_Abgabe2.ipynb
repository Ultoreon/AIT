```python
#Aufg.1
import numpy as np
def my_fib(fib_0=0, fib_1=1, anzahl=10): #Definition der Funktion my_fib()
    a=fib_0                              #übernehmen der Variablen a und b zum rechnen
    b=fib_1
    c=np.zeros(anzahl, dtype=int)        #Erzeugen eines leeren Numpy-Arrays, in das die Lösung übertragen werdden kann
    c[0]=a                               #Eintragen der ersten beiden Werte von fib_0 und fib_1 in das Lösungsarray
    c[1]=b
    for i in range(2,anzahl):            #For-Schleife, in der die Werte der Fibonacci-Reihe übergeben werden
        d=a+b                            #Berechnung des nächsten Elements der Folge
        c[i]=d                           #Übertragen des Ergebnisses
        i+=1                             #inkrementierung der Laufvariablen der Schleife
        b+=a                             #Übernahme des neuen Wertes von b zur Berechnung des nächsten Gliedes der Reihe
        a=b-a                            #Berechnung des neuen Wertes von a (Ich verschiebe quasi nur meine a und b Variablen, damit im nächsten Durchlauf wieder mit ihnen gerechnet werden kann)
    return c
my_fib()                         #Aufruf der Funktion my_fib()

```

```python
#Aufg.2
x, y = np.mgrid[0:5, 0:5]                  #np.mgrid erzeugt ein mehrdimensionales Gitter von Koordinaten.                                     #Im Beispiel wird ein 2D-Gitter von x- und y-Koordinaten für die Bereiche 0 bis 4 erzeugt. 
print('x:\n', x)                           #Die Ausgabe zeigt zwei Arrays, eines für die x-Koordinaten und eines für die y-Koordinaten. 
print('y:\n', y)                           #Dies ist nützlich für die Erstellung von Gitterpunkten für Funktionen oder für die Visualisierung in zwei oder mehr Dimensionen.

def exp2(x):
    return x ** 2

vectorized_function = np.vectorize(exp2)   #np.vectorize transformiert eine Funktion, die auf einzelne Elemente arbeitet, in eine Funktion, die auf NumPy-Arrays arbeitet.
a = np.array([1, 2, 3, 4, 5])              #Dies bedeutet, dass die Funktion auf jedes Element des Arrays angewendet wird. Im Beispiel wird exp2, die ein einzelnes Element quadriert vektorisiert.
Erg = vectorized_function(a)               #Dadurch kann sie auf jedes Element des Arrays [1, 2, 3, 4, 5] angewendet werden, was zu einem neuen Array [1, 4, 9, 16, 25] führt.
print('Ergebnis', Erg)
```

```python
#Aufg.3

def Quersumme(z=1000000):                                                #Definiton der Funktion einer Quersumme                             
    quer=0                                                               #Deklaration von Variablen, die benötigt werden, für 'zahl' wird ein String genutz, da man in diesem eine Referenzierung vornehmen kann
    zahl=str(z)
    for i in range(len(zahl)):                                           #For-Schleife, die ganzen String durchläuft, also die Zahl, von der die Quersumme berechnet werden soll
        quer+=int(zahl[i])                                               #Berechnung der Quersumme
    return quer                                                          #Rückgabe der Quersumme

Anzahl=0                                                                 #Deklaration von Ausgabevariable
for i1 in range(1000000):                                                #For-Schleife zum Durchlaufen aller Zahlen zwischen 0-1000000
    if Quersumme(z=i1)==9:                                               #Überprüfen, ob Quersumme der Zahl 9 ist
        Anzahl+=1                                                        #Wenn ja: Erhöhung von Anzahl
        i1+=1
    else:                                                                #Wenn nein: nächster Schleifendurchlauf
        i1+=1
print('Anzahl aller Zahlen der Quersumme 9 zwischen 0-1000000:', Anzahl) #Ausgabe des Ergebnisses
```

```python
#Aufg.4
def myadvancedfib(anzahl=10,fib1=0,fib2=1,fib3=1,fib4=2):          #Definition der Funktion, der Berechnung der FibonacciReihe von 4 Werten

    fibo={0:fib1,1:fib2,2:fib3, 3:fib4}                           #Definition eines DIctionairies, in wleches die Werte geschrieben werden, hierbei werden
                                                                  #die übergebenen Werte in das Dictionairie geschrieben, dieses bildet somit die ersten Werte der Reihe
    if anzahl in fibo:                                            #Wenn die ANzahl an ELementen in dem Bereich des Dictionairies liegt, wird dieses zurückgegeben
        return fibo
        
    elif anzahl>4:                                                #Wenn die Anzahl an Elementen, die der Grundreihe überschreitet, wird folgende For-Schleife aufgerufen:
        for n in range (4, anzahl+1):                             #Es werden alle Elemente, die außerhalb des Dictionairi sind, neu berechnet
            fibo[n]=fibo[n-1]+fibo[n-2]+fibo[n-3]+fibo[n-4]       #Summe aller vierer vorhergegenagener Ergebnisse
        n+=1                                                      #nächster Shcleifendurchlauf
        return fibo                                               #Rückgabe des neu erzeugten Dictionairis
        
a=myadvancedfib()                                                 #Ausgabe der Reihe
for i in range(0, 10):
    print(a[i])
    i+=1
```

```python
#Aufg.5
import matplotlib as mpl
import matplotlib.pyplot as plt

Dreiwurf = [(x,y,z) for x in range(1,7) for y in range(1,7) for z in range(1,7)]
    #Erzeugung aller möglichen Ergebnisse eines Dreifachwurfes
print('Der Augenzahl des ersten Würfels ist kleiner gleich jener des zweiten, die wiederum kleiner gleich jener des dritten ist')
plt.plot([(x,y,z) for x,y,z in Dreiwurf if x<=y<=z])
    #Filtern aller Kombinationen, mithilfe des <= Opertors
```

```python
print('Die Augenzahl des dritten Würfels ist NICHT gleich der Summe der beiden ersten Würfel')
plt.plot([(x,y,z) for x,y,z in Dreiwurf if (x+y)!=z])
    #Filtern aller möglichen Kombinationen durch addieren von x und y, dann vergleichen, ob das Ergebnis gleich z ist
```

```python
print('Die Gesamtsumme ist kleiner als 12')
plt.plot([(x,y,z) for x,y,z in Dreiwurf if (x+y+z)<12])
    #Filtern alle Kombinationen durch addieren aller drei variablen, dann vergleichen, ob diese kleiner als 12 sind
```

```python
#Aufg.6
#a)
sechser=0
i=0
i2=0
a=(np.random.randint(1,7,40000000))       #Erzeugugng von 40.000.000 Werten zwischen 1-6
b=a.reshape(10000000, 4)                  #Umwandlung in 4er-Tupel, die Wahrscheinlichkeit für jeden Wurf ist gleich, daher macht es keinen unterschied, ob man 4mal 1000000 wirft, oder 4000000
for i in range(9999999):                  #For Schleife zum Zählen aller Tupel, in denen eine 6 vorkommt/ Äußere Schleife durchläuft alle 1000000 4er-Kombis
    i2=0
    for i2 in range(4):                   #Zweite Schleife durchläuft nur die einzelnen Kombinationen
        if b[i][i2]==6:                   #Wenn eine 6 da ist, wird der Zähler inkrementiert und die äußere Schleife geht einen Schritt weiter
            sechser+=1
            i+=1
            break
        else:
            i2+=1
    i+=1
    
print('Wahrscheinlichkeit für bei viermaligem Würfeln wenigstens eine 6 zu erhalten:', (sechser/10000000)*100)
#Ausgabe der Wahrscheinlichkeit, umrechnung in Prozent
```

```python
#b)
ges=0.0
ia=0
pasch=0
for ia in range(1000000):                      #Äußere Schleife, die 1.000.000 mal läuft, um eine große Anzahl an Tests zu bekommen
    
    i=0
    a=(np.random.randint(1,7,48))               #Erzeugugng von 48 Werten zwischen 1-6
    b=a.reshape(24, 2)                          #Umwandlung in 2er-Tupel, die Wahrscheinlichkeit für jeden Wurf ist gleich, daher macht es keinen unterschied, ob man 2mal 24 wirft, oder 48 mal
    
    i=0
    for i in range(24):                         #For Schleife zum Zählen aller Tupel, in denen ein 6er-Pasch vorkommt; Schleife durchläuft alle 24 2er-Kombis
        if b[i][0]==b[i][1]==6:                 #Wenn ein 6er-Pasch da ist, wird der Zähler(pasch) inkrementiert und die Schleife geht einen Schritt weiter
            pasch+=1
            ia+=1
            break
        else:                                   #Wenn Tupel kein 6er-Pasch, innere Schleife um 1 weiter gehen
            i+=1  
    ia+=1
print((pasch/1000000)*100)                      #Ausgabe der Wahrscheinlichkeit, umrechnung in Prozent
```

```python
#Aufg.7
lo=range(1,50)                                            #Erzeugung der Zahlen 1-49
ges=0
p=0
h=1                                                       #Angabe, wie oft die äußerste Schleife durchlaufen soll, um Mittelwert zu erhalten, ich würde nichts über 5 Empfehlen, dauert sonst sehr sehr lange
for p in range(h):                                        #Äußerste Schleife, damit Berechnung des Wertes öfters durchgeführt werden kann
    b = np.random.choice(lo,size=6, replace=False)        #erzeugen von b als erstes Glied aller Lottoergebnisse
    b.sort()                                              #sortieren, damit vergleichen einfacher ist
    abbr=0
    for n in range(5000):                                 #Erste innere Schleife, die bis 5000 läuft, in Aufgabenstellung war 3016 als erste Wiederholung angegeben, mit Reserve dann 5000
        a = np.random.choice(lo,size=6, replace=False)    #erzeugen einer neuen Lotto-Ziehung
        a.sort()                                    
        i=0
        for i in range(len(b)):                           #innerste Schleife, in der die Ergebnisse verglichen werden
            if np.array_equal(a,b[i]):
                ges+=i                                    #Aufaddieren aller Ergebnisse, damit naachher mittelwert gebildet werden kann
                abbr=1
                break   
            else:
                i+=1
        b=np.vstack([a,b])                                #Aufkummulieren aller bisherigen Ziehungen, damit verglichen werden kann
        if abbr==1:
            break
    p+=1
print(ges/h)                                             #Ausgabe des Ergebnisses, als Mittelwert


    
```
