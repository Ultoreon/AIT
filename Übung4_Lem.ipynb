## Aufgabe 1
Führen Sie in Python numerische Zufallsexperimente mit dem Doppelwürfel durch.

**Frage:** Wie viele Doppelwürfe dauert es (im Schnitt), bis Sie einen 6er Pasch „würfeln“?

Stellen Sie das Ergebnis in einem Scatter-Diagramm dar!

````python
import random
import numpy as np
import matplotlib.pyplot as plt

def ein_6er_pasch():
    j = 0
    while True:
        a = random.randint(1,6)
        b = random.randint(1,6)
        j += 1
        if  a == 6 and b == 6:
           break 
    return j

def durchschnitt_6er_pasch(n):
    j = 0
    Liste = []
    for i in range(n):
        x = ein_6er_pasch()
        Liste.append(x)
        j += x
    return Liste

def plot_durschnitt_6er_pasch(n):
    x = np.arange(1, n+1)
    y = durchschnitt_6er_pasch(n)
    plt.scatter(x, y, label='Einzelne Würfe', color='blue', alpha=0.5)
    plt.xlabel('Anzahl der Würfe')
    plt.ylabel('Durchschnitt')
    plt.title('Durchschnitt der Würfe bis zum 6er Pasch')
    plt.legend()
    plt.grid()
    plt.show()

plot_durschnitt_6er_pasch(1000)
````

Vorgegeben seien die Punkte:
```
(1,7), (2,6), (3,4), (4, 5), (5, 3), (6,5), (7,7), (9,8), (10,5)
```

Stellen Sie diese Punkte in einem `matplotlib`-Diagramm dar und finden Sie jeweils mit **BSplines** und **Lagrange-Interpolation** kontinuierliche Funktionen, die diese Punkte verbinden.

Geben Sie zusätzlich eine andere, selbst gewählte Punktmenge an und vergleichen Sie die jeweiligen Ergebnisse der Interpolation!

````python

import numpy as np
import matplotlib.pyplot as plt
import scipy.interpolate as si

punkte = [(1,7), (2,6), (3,4), (4, 5), (5, 3), (6,5), (7,7), (9,8), (10,5)]
def plot_spline_interpolation(punkte):
    x, y = zip(*punkte)
    x = np.array(x)
    y = np.array(y)

    spl = si.splrep(x, y)
    x1 = np.linspace(0, 10, 200)
    y1 = si.splev(x1, spl)

    lagrange_polynom = si.lagrange(x,y)
    x3 = np.linspace(0, 10, 200)
    y3 = lagrange_polynom(x3)

    plt.plot(x, y, 'ro', label='Einzelne Punkte')
    plt.plot(x1, y1, 'b-', label='Spline-Interpolation')
    plt.plot(x3, y3, 'g-', label='Lagrange-Interpolation')
    plt.ylim(-5, 20)
    plt.xlabel('x-Werte')
    plt.ylabel('y-Werte')
    plt.title('Splines und Interpolation')
    plt.style.use("classic")
    plt.legend()
    plt.grid()
    plt.show()
plot_spline_interpolation(punkte)
````

````python
punkte2 = [(1,2), (2,8), (3,3), (4, 4), (5, 0), (6,9), (7,7), (8,9)]

plot_spline_interpolation(punkte2)
````

\[ z = f(x, y) = x^2 - y^2 \]

## Aufgabe 5
Geben Sie die vollständige **Konfusionsmatrix** sowie die in der Vorlesung behandelten Aggregatwerte **Recall, Precision, Accuracy, F1-Score, Sensitivität** und **Spezifität** an für folgende Beispiele:

**a)** Diagnosealgorithmus mit 1342 Fällen:
- 876 positive Vorhersagen → davon 188 falsch (tatsächlich negativ)
- Restliche negative Vorhersagen → 123 falsch (tatsächlich positiv)

**b)** Algorithmus sagt immer „positiv“ bei 100 Fällen → tatsächlich nur 75 positiv

**c)** Algorithmus rät zufällig mit 50% „positiv“ / „negativ“ bei 100 Datensätzen, davon 50% tatsächlich positiv

**d)** Wie ändern sich die Werte aus **c)**, wenn:
- Der Algorithmus zu 90% „positiv“ rät und
  - die reale Positivrate **90%** beträgt?
  - die reale Positivrate **10%** beträgt?

````python
def Konfusionsmatrix(FP, FN, TN, TP):
    matrix = np.array([[TN, FP], [FN, TP]])
    return matrix

def Aggregatwerte(FP, FN, TN, TP):
    accuracy = round((TP + TN) / (TP + TN + FP + FN),4)
    precision = round(TP / (TP + FP) ,4)
    recall = round(TP / (TP + FN) ,4)
    f1_score = round((2 *TP)/((2*TP) + FN + FP) ,4)
    Sensitivity = recall
    Specificity = round(TN / (TN + FP) if (TN + FP) != 0 else 0,4)
    return accuracy, precision, recall, f1_score , Sensitivity, Specificity
````

````python
#a

print('Konfusionsmatrix:\n', Konfusionsmatrix(FP=188, FN=123, TN=155, TP=688))
accuracy_val, precision_val, recall_val, f1_score_val, sensitivity_val, specificity_val = Aggregatwerte(FP=188, FN=123, TN=155, TP=688)
print(f'Accuracy: {accuracy_val}', f' Precision: {precision_val}', f' Recall: {recall_val}', f' F1_Score: {f1_score_val}\n', f' Sensitivity: {sensitivity_val}', f' Specificity: {specificity_val}')
````

````python
#b
print('Konfusionsmatrix:\n', Konfusionsmatrix(FP=25, FN=0, TN=0, TP=75))
accuracy_val, precision_val, recall_val, f1_score_val, sensitivity_val, specificity_val = Aggregatwerte(FP=25, FN=0, TN=0, TP=75)
print(f'Accuracy: {accuracy_val}', f' Precision: {precision_val}', f' Recall: {recall_val}', f' F1_Score: {f1_score_val}\n', f' Sensitivity: {sensitivity_val}', f' Specificity: {specificity_val}')
````

## Aufgabe 6 (freiwillig)
Schreiben Sie ein (rekursives) Python-Programm zur Lösung des Spiels **„Towers of Hanoi“**:

Ziel: Alle Scheiben von **Turm A** nach **Turm B** verschieben, dabei gilt:
- Immer nur eine Scheibe auf einmal bewegen
- Niemals eine größere auf eine kleinere legen
- **Turm C** dient als Zwischenlager

Ihr Programm sollte `solve_towers_of_hanoi` heißen und 4 Parameter akzeptieren:
- Anzahl der Scheiben
- Namen der Türme (Start, Ziel, Zwischenlager)

*Hinweis: Eine Lösung mit nur 5 Zeilen im Rumpf der Funktion ist möglich!*

````python
def solve_towers_of_hanoi(n, source, target, auxiliary):
    if n == 1:
        print(f"Bewege Scheibe 1 von {source} nach {target}")
        return
    solve_towers_of_hanoi(n - 1, source, auxiliary, target)
    print(f"Bewege Scheibe {n} von {source} nach {target}")
    solve_towers_of_hanoi(n - 1, auxiliary, target, source)

def towers_of_hanoi(n):
    print(f"Anzahl der Scheiben: {n}")
    solve_towers_of_hanoi(n, 'A', 'C', 'B')
    
towers_of_hanoi(3)  # Beispiel mit 3 Scheiben
````
